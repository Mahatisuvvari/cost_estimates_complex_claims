#### Handling Missing Values

```python
# Handle missing values in 'MaritalStatus' by filling with the mode
df['MaritalStatus'].fillna(df['MaritalStatus'].mode()[0], inplace=True)

# Convert date columns to datetime
df['DateTimeOfAccident'] = pd.to_datetime(df['DateTimeOfAccident'])
df['DateReported'] = pd.to_datetime(df['DateReported'])

# Extract useful features from date columns
df['AccidentYear'] = df['DateTimeOfAccident'].dt.year
df['AccidentMonth'] = df['DateTimeOfAccident'].dt.month
df['ReportYear'] = df['DateReported'].dt.year
df['ReportMonth'] = df['DateReported'].dt.month

# Display the first few rows to verify changes
df.head()
```

#### Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the plotting environment
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(14, 10))

# Continuous features distributions
continuous_features = ['Age', 'WeeklyWages', 'HoursWorkedPerWeek', 'InitialIncurredClaimsCost', 'UltimateIncurredClaimCost']
for i, feature in enumerate(continuous_features, 1):
    plt.subplot(3, 2, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

The distributions of the continuous features are as follows:

**Age**: The age distribution appears to be roughly normal, centered around the mid-30s.

**WeeklyWages**: This distribution is right-skewed, with most wages clustering around the lower end but with some high outliers.

**HoursWorkedPerWeek**: Most claims involve standard work hours (around 38-40 hours), but there are notable outliers.

**InitialIncurredClaimsCost**: This is highly right-skewed, indicating that while most claims have relatively low initial costs, a few claims have significantly higher costs.

**UltimateIncurredClaimCost**: Similar to initial costs, this feature is right-skewed with most claims having lower costs and a few outliers with very high costs.

```python
# Categorical features distributions
categorical_features = ['Gender', 'MaritalStatus', 'PartTimeFullTime', 'DaysWorkedPerWeek']

plt.figure(figsize=(14, 8))

for i, feature in enumerate(categorical_features, 1):
    plt.subplot(2, 2, i)
    sns.countplot(data=df, x=feature, palette='viridis')
    plt.title(f'Distribution of {feature}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

The distributions of the categorical features are as follows:

**Gender**: There are more male claimants than female claimants.
    
**MaritalStatus**: Most claimants are married, with smaller proportions of single and unknown marital status.
    
**PartTimeFullTime**: The majority of the claimants work full-time.
    
**DaysWorkedPerWeek**: Most claimants work five days per week, with some working fewer or more days.

### Feature Engineering.

```python
# Feature Engineering: Extracting useful information from date columns

# Convert date columns to datetime format
df['DateTimeOfAccident'] = pd.to_datetime(df['DateTimeOfAccident'])
df['DateReported'] = pd.to_datetime(df['DateReported'])

# Create new features
df['AccidentYear'] = df['DateTimeOfAccident'].dt.year
df['AccidentMonth'] = df['DateTimeOfAccident'].dt.month
df['ReportYear'] = df['DateReported'].dt.year
df['ReportMonth'] = df['DateReported'].dt.month
df['ReportDelayDays'] = (df['DateReported'] - df['DateTimeOfAccident']).dt.days

# Display the first few rows with the new features
df.head()
```

```python
df.head()
```

```python
data = df.copy()

# Extract year and month from datetime columns for analysis
data['AccidentYear'] = data['DateTimeOfAccident'].dt.year
data['AccidentMonth'] = data['DateTimeOfAccident'].dt.month
data['ReportedYear'] = data['DateReported'].dt.year
data['ReportedMonth'] = data['DateReported'].dt.month

# Plot the number of claims per year
plt.figure(figsize=(12, 6))
sns.countplot(x='AccidentYear', data=data, palette='viridis')
plt.title('Number of Claims per Accident Year')
plt.xlabel('Accident Year')
plt.ylabel('Number of Claims')
plt.xticks(rotation=45)
plt.show()

# Plot the number of claims reported per year
plt.figure(figsize=(12, 6))
sns.countplot(x='ReportedYear', data=data, palette='viridis')
plt.title('Number of Claims Reported per Year')
plt.xlabel('Reported Year')
plt.ylabel('Number of Claims')
plt.xticks(rotation=45)
plt.show()

# Plot the number of claims per month
plt.figure(figsize=(12, 6))
sns.countplot(x='AccidentMonth', data=data, palette='viridis')
plt.title('Number of Claims per Accident Month')
plt.xlabel('Accident Month')
plt.ylabel('Number of Claims')
plt.show()

# Plot the number of claims reported per month
plt.figure(figsize=(12, 6))
sns.countplot(x='ReportedMonth', data=data, palette='viridis')
plt.title('Number of Claims Reported per Month')
plt.xlabel('Reported Month')
plt.ylabel('Number of Claims')
plt.show()
```

**Time-Related Trends in Claims**

**Number of Claims per Accident Year**
The count of claims per accident year shows variations over time. Certain years have significantly higher numbers of claims, indicating possible factors like regulatory changes, industry-specific events, or economic conditions.

**Number of Claims Reported per Year**
The number of claims reported per year follows a similar pattern to the accident year, with some delays between the accident year and the reported year. This might reflect the time taken by claimants to report accidents.

**Number of Claims per Accident Month**
Claims are distributed relatively evenly across months, but certain months show higher frequencies. This might be linked to seasonal employment patterns, weather conditions, or specific industry activities.

**Number of Claims Reported per Month**
The distribution of claims reported per month also exhibits seasonal trends. Peaks in reporting might coincide with specific business cycles or periods of increased activity in relevant industries.

```python
# Analyze cost trends over time

# Average claim cost per accident year
avg_cost_per_accident_year = data.groupby('AccidentYear')['UltimateIncurredClaimCost'].mean().reset_index()

# Plot average claim cost per accident year
plt.figure(figsize=(12, 6))
sns.lineplot(x='AccidentYear', y='UltimateIncurredClaimCost', data=avg_cost_per_accident_year, marker='o')
plt.title('Average Ultimate Incurred Claim Cost per Accident Year')
plt.xlabel('Accident Year')
plt.ylabel('Average Ultimate Incurred Claim Cost')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Average claim cost per reported year
avg_cost_per_reported_year = data.groupby('ReportedYear')['UltimateIncurredClaimCost'].mean().reset_index()

# Plot average claim cost per reported year
plt.figure(figsize=(12, 6))
sns.lineplot(x='ReportedYear', y='UltimateIncurredClaimCost', data=avg_cost_per_reported_year, marker='o')
plt.title('Average Ultimate Incurred Claim Cost per Reported Year')
plt.xlabel('Reported Year')
plt.ylabel('Average Ultimate Incurred Claim Cost')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Average claim cost per accident month
avg_cost_per_accident_month = data.groupby('AccidentMonth')['UltimateIncurredClaimCost'].mean().reset_index()

# Plot average claim cost per accident month
plt.figure(figsize=(12, 6))
sns.lineplot(x='AccidentMonth', y='UltimateIncurredClaimCost', data=avg_cost_per_accident_month, marker='o')
plt.title('Average Ultimate Incurred Claim Cost per Accident Month')
plt.xlabel('Accident Month')
plt.ylabel('Average Ultimate Incurred Claim Cost')
plt.grid(True)
plt.show()

# Average claim cost per reported month
avg_cost_per_reported_month = data.groupby('ReportedMonth')['UltimateIncurredClaimCost'].mean().reset_index()

# Plot average claim cost per reported month
plt.figure(figsize=(12, 6))
sns.lineplot(x='ReportedMonth', y='UltimateIncurredClaimCost', data=avg_cost_per_reported_month, marker='o')
plt.title('Average Ultimate Incurred Claim Cost per Reported Month')
plt.xlabel('Reported Month')
plt.ylabel('Average Ultimate Incurred Claim Cost')
plt.grid(True)
plt.show()
```

#### Cost Trends Over Time

**Average Ultimate Incurred Claim Cost per Accident Year**
The average claim cost per accident year shows fluctuations, indicating possible economic or industry-specific factors influencing claim costs. Certain years exhibit higher average costs, which may correlate with significant events or policy changes.

**Average Ultimate Incurred Claim Cost per Reported Year**
The trend of average claim costs per reported year mirrors the accident year trends, with some lag. This aligns with the time taken for claims to be reported and processed.

**Average Ultimate Incurred Claim Cost per Accident Month**
Monthly trends show variability, with some months experiencing higher average claim costs. This could be due to seasonal employment patterns, weather-related incidents, or other time-specific factors.

**Average Ultimate Incurred Claim Cost per Reported Month**
The monthly reporting trends also show fluctuations, which might be influenced by the same factors affecting the accident month trends.

#### Complex Claims identification

```python
# Define complex claims as those in the top 10% of UltimateIncurredClaimCost
threshold = df['UltimateIncurredClaimCost'].quantile(0.90)
df['ComplexClaim'] = (df['UltimateIncurredClaimCost'] >= threshold).astype(int)
```

```python
df[['Age', 'DependentChildren', 'DependentsOther', 'WeeklyWages', 'HoursWorkedPerWeek', 'DaysWorkedPerWeek',
       'ClaimDescriptionKeyword_0', 'ClaimDescriptionKeyword_1',
       'ClaimDescriptionKeyword_2', 'ClaimDescriptionKeyword_3',
       'ClaimDescriptionKeyword_4', 'ClaimDescriptionKeyword_5',
       'ClaimDescriptionKeyword_6', 'ClaimDescriptionKeyword_7',
       'ClaimDescriptionKeyword_8', 'ClaimDescriptionKeyword_9',
       'ClaimDescriptionKeyword_10', 'ClaimDescriptionKeyword_11',
       'InitialIncurredClaimsCost', 'UltimateIncurredClaimCost',
       'AccidentYear', 'AccidentMonth', 'ReportYear', 'ReportMonth',
       'ReportDelayDays', 'ComplexClaim']].groupby('ComplexClaim').mean()
```

#### Insights from Complex Claims Analysis:
    
**Age:** Complex claims tend to involve older claimants (mean age ~38) compared to non-complex claims (mean age ~33).

**WeeklyWages:** Higher weekly wages are associated with complex claims ($565) versus non-complex claims ($400).

**Initial and Ultimate Incurred Claims Cost:** As expected, initial and ultimate costs are significantly higher for complex claims.

**Report Delay:** Complex claims have a longer delay between the accident and the report (~47 days vs. ~37 days).

#### Identifying Correlation

```python
# Correlation Analysis: Correlation matrix and heatmap

# Compute correlation matrix
correlation_matrix = df[['Age', 'DependentChildren', 'DependentsOther', 'WeeklyWages', 'HoursWorkedPerWeek', 'DaysWorkedPerWeek',
       'ClaimDescriptionKeyword_0', 'ClaimDescriptionKeyword_1',
       'ClaimDescriptionKeyword_2', 'ClaimDescriptionKeyword_3',
       'ClaimDescriptionKeyword_4', 'ClaimDescriptionKeyword_5',
       'ClaimDescriptionKeyword_6', 'ClaimDescriptionKeyword_7',
       'ClaimDescriptionKeyword_8', 'ClaimDescriptionKeyword_9',
       'ClaimDescriptionKeyword_10', 'ClaimDescriptionKeyword_11',
       'InitialIncurredClaimsCost', 'UltimateIncurredClaimCost',
       'AccidentYear', 'AccidentMonth', 'ReportYear', 'ReportMonth',
       'ReportDelayDays', 'ComplexClaim']].corr()

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Insights from Correlation Analysis:
    
**Strong Correlations:**

**InitialIncurredClaimsCost and UltimateIncurredClaimCost:** A strong positive correlation, indicating that initial estimates are predictive of final costs.

**WeeklyWages and HoursWorkedPerWeek:** Moderate positive correlation, as higher wages often correspond to more hours worked.

**Moderate Correlations:**

**Age and Claim Costs:** Age shows some correlation with both initial and ultimate claim costs.

**ReportDelayDays and Claim Costs:** Longer delays in reporting are moderately correlated with higher claim costs, indicating potential complexity.
